```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
dataMax = 200
dataSize = 50
noise = 50

def create_data():
    #real equation: y = 0.1x + 2
    data = np.arange(0,dataMax,dataMax/dataSize)
    return data

def randomise(x):
    y = []
    for i in range(np.size(x)):
        y.append(x[i] + random.uniform(-noise,noise))
    return y
    
def draw_random_data():
    dataSet = create_data()
    noisyData = randomise(dataSet)
    plt.plot(dataSet,"r",label = "clean data")
    plt.plot(noisyData,"bo",label = "noisy data")
    plt.legend()
    plt.show()
    return noisyData
    
noisyData = draw_random_data()
X = np.arange(0,dataSize,1)
```

```python
#linear regression equation with one independant variable: yi = bxi
#for simplicity in this case that the y intercept is at 0
#error equation: e = (yi - bxi)^2
#error differential: de/db = -2xi(yi - bxi)
def regression(X,Y):
    learningRate = 0.0001
    iterations = 100
    n = np.size(Y)
    b = random.uniform(0,1)
    print(b)
    for i in range(iterations):
        b -= (cost_derivative(n,b,X,Y)/float(n)) * learningRate
        print(b)
    draw_prediction(n,b,X)

def cost_derivative(n,b,X,Y):
    bDeriv = 0
    for i in range(n):
        bDeriv += -2 * X[i] * (Y[i] - b * X[i])
    return bDeriv

def draw_prediction(n,b,X):
    Y = []
    for i in range(n):
        Y.append(X[i] * b)
    plt.plot(X,Y,"g")
    plt.plot(noisyData,"bo")
    plt.plot(create_data(),"r")

regression(X,noisyData)
```
