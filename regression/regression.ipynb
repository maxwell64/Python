```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
dataMin = 127
dataMax = 194
dataSize = 100
noise = 20

def create_data():
    data = np.arange(dataMin,dataMax,(dataMax - dataMin)/dataSize)
    return data

def randomise(x):
    y = []
    for i in range(np.size(x)):
        y.append(x[i] + random.uniform(-noise,noise))
    return y
    
def draw_random_data():
    dataSet = create_data()
    noisyData = randomise(dataSet)
    plt.plot(dataSet,"r",label = "clean data")
    plt.plot(noisyData,"bo",label = "noisy data")
    plt.legend()
    plt.show()
    return noisyData
    
noisyData = draw_random_data()
X = np.arange(0,dataMax - dataMin,(dataMax - dataMin)/dataSize)
```

```python
#linear regression equation with one independant variable: yi = bxi + c
#error equation: e = (yi - bxi - c)^2
#error differential: de/db = -2xi(yi - bxi - c) and de/dc = -2(yi - bxi - c)
def regression(X,Y):
    learningRate = 0.0001
    iterations = 5000
    n = np.size(X)
    b = random.uniform(0,1)
    c = random.uniform(0,100)
    print(b,c)
    for i in range(iterations):
        b -= (cost_derivative(n,b,c,X,Y)[0]/float(n)) * learningRate
        c -= (cost_derivative(n,b,c,X,Y)[1]/float(n)) * learningRate * 100
    draw_prediction(n,b,c,X)
    print(b,c)

def cost_derivative(n,b,c,X,Y):
    bDeriv = 0
    cDeriv = 0
    for i in range(n):
        bDeriv += -2 * X[i] * (Y[i] - b * X[i] - c)
        cDeriv += -2 * (Y[i] - b * X[i] - c)
    return [bDeriv,cDeriv]

def draw_prediction(n,b,c,X):
    Y = []
    for i in range(n):
        Y.append(X[i] * b + c)
    plt.plot(Y,"g")
    plt.plot(noisyData,"bo")
    plt.plot(create_data(),"r")

regression(X,noisyData)
```
