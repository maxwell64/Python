```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
def createData():
    #real equation: y = 0.1x + 2
    data = np.arange(0,10,0.1)
    return data

def randomise(x):
    y = []
    noise = 2
    for i in range(np.size(x)):
        y.append(x[i] + random.uniform(-noise,noise))
    return y
    
def drawRandomData():
    dataSet = createData()
    noisyData = randomise(dataSet)
    plt.plot(dataSet,"r",label = "clean data")
    plt.plot(noisyData,"bo",label = "noisy data")
    plt.legend()
    plt.show()
    return noisyData
    
noisyData = drawRandomData()
```
