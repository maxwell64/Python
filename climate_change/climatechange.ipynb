Azure Notebook Climate Change Analysis

```python
import matplotlib.pyplot as graph
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns; sns.set()
```

Import Data

```python
yearsBase, meanBase = np.loadtxt('5-year-mean-1951-1980.csv', delimiter=',', usecols=(0, 1), unpack=True)
years, mean = np.loadtxt('5-year-mean-1882-2014.csv', delimiter=',', usecols=(0, 1), unpack=True)
```

Create a scatter plot

```python
graph.scatter(yearsBase, meanBase)
graph.title('scatter plot of mean temp difference vs year')
graph.xlabel('years', fontsize=12)
graph.ylabel('mean temp difference', fontsize=12)
graph.show
```

Perform linear regression

```python
# Creates a linear regression from the data points
m,b = np.polyfit(yearsBase, meanBase, 1)

# Simple mx + b linea function
def f(x):
    return m*x + b

# Same graph as above but adds the trendline
graph.scatter(yearsBase, meanBase)
graph.plot(yearsBase, f(yearsBase))
graph.title('scatter plot of mean temp difference vs year')
graph.xlabel('years', fontsize=12)
graph.ylabel('mean temp difference', fontsize=12)

# Prints the computed equation of the trendline
print('y = {0}x + {1}'.format(m,b))
graph.show()
```

Perform linear regression with scikit-learn

```python
# Pick and instance the linear regression model
model = LinearRegression(fit_intercept=True)

# Fit/build model
model.fit(yearsBase[:, np.newaxis], meanBase)
mean_predicted = model.predict(yearsBase[:, np.newaxis])

# Generate plot as previous
graph.scatter(yearsBase, meanBase)
graph.plot(yearsBase, mean_predicted)
graph.title('scatter plot of mean temp difference vs year')
graph.xlabel('year', fontsize=12)
graph.ylabel('mean temp difference', fontsize=12)
graph.show()

# Prints the equation of the trendline
print('y = {0}x + {1}'.format(model.coef_[0], model.intercept_))
```

Perform linear regression with Seaborn

```python
graph.scatter(years, mean)
graph.title('scatter plot of mean temp difference vs year')
graph.xlabel('years', fontsize=12)
graph.ylabel('mean temp difference', fontsize=12)
sns.regplot(yearsBase, meanBase)
graph.show()
```
