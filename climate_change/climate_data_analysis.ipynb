Import modules and dataset

```python
print('importing modules and data...', end=' ')
import numpy as np
import matplotlib.pyplot as graph
%matplotlib inline
graph.rcParams['figure.figsize']=(10,5)
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
yearsBase, meanBase = np.loadtxt('5-year-mean-1951-1980.csv', delimiter=',', usecols=(0,1), unpack=True)
years, mean = np.loadtxt('5-year-mean-1882-2014.csv', delimiter=',', usecols=(0,1), unpack=True)
print('done')
```

Scatter plot of the data

```python
graph.scatter(yearsBase, meanBase)
graph.xlabel('year', fontsize=12)
graph.ylabel('mean temperature difference', fontsize=12)
graph.title('mean temperature difference vs year')
graph.show()
```

Perform linear regression on the data using numpy

```python
m, b = np.polyfit(yearsBase, meanBase, 1)

def f(x):
    return m*x + b

graph.scatter(yearsBase, meanBase)
graph.plot(yearsBase, f(yearsBase), 'r')
graph.xlabel('year', fontsize=12)
graph.ylabel('mean temperature difference', fontsize=12)
graph.title('mean temperature difference vs year')
graph.show()

print('y = {0}x + {1}'.format(m, b))
```

```python
n, c = np.polyfit(years, mean, 1)

def g(x):
    return n*x + c

graph.scatter(years, mean)
graph.plot(years, g(years), 'r')
graph.xlabel('year', fontsize=12)
graph.ylabel('mean temperature difference', fontsize=12)
graph.title('mean temperature difference vs year')
graph.show()

print('y = {0}x + {1}'.format(n, c))
```

Linear regression with scikit-learn

```python
model = LinearRegression(fit_intercept=True)

model.fit(yearsBase[:, np.newaxis], meanBase)
mean_predicted = model.predict(yearsBase[:, np.newaxis])

graph.scatter(yearsBase, meanBase)
graph.plot(yearsBase, mean_predicted, 'r')
graph.xlabel('year', fontsize=12)
graph.ylabel('mean temperature difference', fontsize=12)
graph.title('mean temperature difference vs year')
graph.show()

print('y = {0}x + {1}'.format(model.coef_[0], model.intercept_))
```

```python
model = LinearRegression(fit_intercept=True)

model.fit(years[:, np.newaxis], mean)
mean_predicted = model.predict(years[:, np.newaxis])

graph.scatter(years, mean, s=20)
graph.plot(years, mean_predicted, 'r')
graph.xlabel('year', fontsize=12)
graph.ylabel('mean temperature difference', fontsize=12)
graph.title('mean temperature difference vs year')
graph.show()

print('y = {0}x + {1}'.format(model.coef_[0], model.intercept_))
```

Regression analysis with seaborn

```python
graph.scatter(years, mean, s=10)
graph.xlabel('year', fontsize=12)
graph.ylabel('mean temperature difference', fontsize=12)
graph.title('mean temperature difference vs year')
sns.regplot(yearsBase, meanBase, scatter=False, color='red')
graph.show
```
